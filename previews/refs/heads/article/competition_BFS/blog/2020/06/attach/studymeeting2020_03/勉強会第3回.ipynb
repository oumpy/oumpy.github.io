# for文やif文の組み合わせ
for文は処理を繰り返す文で、
```
for i in range(10):
  (処理)
```
のように書きます。
実はこの(処理)の中にもif文やfor文,while文を書くことができます。

for文の中にfor文を書く場合、ループ変数を別のものにしなければいけないことに注意してください。

同様に、if文の中にfor文やif文を入れることもできます。やりすぎるとコードが汚くなるので注意です！(4重ループ以上はほぼ見ません)

````python
for i in range(5):
  for j in range(5):
    print(i,j)
````

````python
for i in range(1,101):
  if i%15 == 0:
    print("FizzBuzz")
  elif i%3 == 0:
    print("Fizz")
  elif i%5 == 0:
    print("Buzz")
  else:
    print(i)
````

### 問題1
2つのリストL,Rが与えられます。
L,Rの中身はすべて整数です。L,Rのうちからそれぞれ1つずつ整数をとってきて和をとります。これが100**以上**となる組はいくつあるでしょうか？

````python
L=[116, 52, 105, 29, 45, 69, 84, 126, 80, 83, 62, 133]
R=[-30, 32, 13, -7, -41, 4, 38, 43, -20, 69, 23, -27]
#L,Rの要素数は12です。
ans = 0
for i in range(12):
  for j in range(12):
    if L[i] + R[j] >= 100:
      ans = ans+1
print(ans)
````

### 問題1の答え

````python
#二重ループを使います。
L=[116, 52, 105, 29, 45, 69, 84, 126, 80, 83, 62, 133]
R=[-30, 32, 13, -7, -41, 4, 38, 43, -20, 69, 23, -27]
ans=0
for i in range(12):
  for j in range(12):
    if L[i] + R[j] >= 100:
      ans = ans + 1 #条件をみたしていれば答えを1増やす
print(ans)
````

このように、あり得るパターンをすべて調べ上げる手法のことを**全探索**といいます。この問題では組の数は144通りしかないので短時間ですべて調べ上げることができます。

# ライブラリ、組み込み関数について
print()やlen()のように、何もしなくても使える関数のことを**組み込み関数**といいます。これらのほかに、モジュールを複数集めたパッケージの集合体である**ライブラリ**と呼ばれるものが存在します。モジュールとは**一連の作業を実行するためのファイル**のことで、それを集めたものがパッケージです。ライブラリはパッケージをまとめてインストールできるようにしたものです。（細かい用語の違いは覚えなくてOKです！）

ライブラリを使うにはまずimportをする必要があります。例えばmathというライブラリを使いたいなら
```
import math
```
のように書きます。複数のライブラリをimportしたい場合は複数行にわたってインポートを行えばよいです。
ライブラリは競技プログラミングに限らず、ほかの分野でもよく出てくるものなので使い方をマスターしましょう！

# 計算量について 
今まで練習問題として挙げてきたものは、どれも処理するデータ数が少なく、実行時間を気にすることなくコードを書けるものばかりでした。しかし、問題の難易度が上がってくると処理するべきデータ数が多くなってくるので、工夫してよりよいアルゴリズムを用いて問題の答えを出さないといけません。競技プログラミングにおいては処理が重すぎるプログラムを提出すると**TLE(Time Limit Exceeded,実行時間制限超過)**という不正解判定を食らいます。

計算量を減らさないといけない問題の例を見てみましょう：
[Sum of Three Integers(ABC051-B)](https://atcoder.jp/contests/abc051/tasks/abc051_b)

### TLEになるコードの例
先ほどの多重ループを活用して、次のようなコードを書きたくなったかもしれません。

````python
K,S = map(int,input().split())
ans = 0
for X in range(0,K+1):
  for Y in range(0,K+1):
    for Z in range(0,K+1):
      if X + Y + Z == S:
        ans += 1
print(ans)
````

しかし、このコードを提出すると**TLE**になります。($K,S$を大きい数にして試してみると実行に時間がかかっているのがわかると思います)

このコードにおいて、計算量は$(K+1)^3$です。一番大きい項に近似して、オーダーを$O(K^3)$といいます。制約は$2 \le K \le 2500$ なので、計算量は最高で$2500^3 \fallingdotseq 1.56 \times 10^{10} $ほどになります。
およそ1秒間に処理できる回数は$10^6 \sim 10^7$程度なので、制限時間の2秒には間に合いません。

ではどのようにすれば計算量を落とせるか考えてみましょう。

### AC(正解)のコード例

$X+Y+Z=S$に注目します。$S$は定数なので、$X,Y$を決めてしまえば$Z=S-X-Y$と決まります。つまり、$X,Y$を全探索して、$Z$が$0 \le Z \le K$の範囲に入るかを見てやればOKです。この考え方であれば、計算量は$2500^2 \fallingdotseq 6.2 \times 10^6$となり間に合います。

````python
K,S=map(int,input().split())
ans=0
for X in range(0,K+1):
  for Y in range(0,K+1):
    Z=S-X-Y
    if 0 <= Z <= K:
      ans+=1
print(ans)
````

このように、少し考察することで計算量のオーダーを$O(K^2)$に落とすことができました！計算量を落とすために、数学的な考察をすることのほかにも、さまざまなアルゴリズムを用いることがあります。これらに触れると競プロがもっと楽しくなること間違いなしです！\

### 他の例
考察することで計算量が落とす(高速化する)ことは競プロの醍醐味の一つです。ここではいくつか例を挙げます。

1.   ある整数Nの約数の個数を数える\
愚直に1からNまで割り切れるか試すと計算量が$O(N)$となりますが、実際は調べるのは1から$ \sqrt N $を超えない最大の整数までで十分なので、計算量は$O(\sqrt N)$です。よって$10^{12}$程度の数なら約数の個数を2秒以内に数えきることができます。
