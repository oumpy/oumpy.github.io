## NumPyで数値計算 

```python
import numpy as np
```

```python
x = np.arange(0, 1, 0.1)
print('x:', x)
```

```python
y = 2*x + 10
print('y:', y)
```

## Matplotlibで図を描画する

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x, y)
plt.show()
```

## Seabornによる回帰直線の描画

```python
import seaborn as sns
```

テストデータ($y=2x+10+\xi$)を作成します。ただし、$x$と$\xi$は正規分布からサンプリングされたとします。データ数は`n_data`です。

```python
np.random.seed(0) # 乱数seedの設定

# Toyデータ(y=2*x+10+noise)の作成
n_data = 100 # データ数
x = np.random.randn(n_data)
y = 2*x + np.random.randn(n_data) + 10
```

```python
# 回帰直線のplot 
plt.figure(figsize=(4,4))
sns.regplot(x, y) # 回帰(regression)の実行
plt.show() # 画像表示
```
